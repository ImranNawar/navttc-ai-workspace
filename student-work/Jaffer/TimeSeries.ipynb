```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
print(flights.head())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=flights, x='year', y='passengers', hue='month', palette='tab10')

plt.title('Monthly Number of Passengers Over Years', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Passengers', fontsize=14)
plt.legend(title='Month', title_fontsize='13', fontsize='11')
# plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='year', y='passengers', data=flights, estimator=sum, palette='tab10')
plt.title('Total passengers per year',fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14, fontstyle='italic')
plt.ylabel('Total passengers' ,fontsize=14, fontstyle='italic')
plt.show()
```
